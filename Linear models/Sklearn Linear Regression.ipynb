# **Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patheffects as pe
import seaborn as sns
import sklearn

import warnings
warnings.filterwarnings('ignore')
```

## **plt settings**

```python
plt.style.use('default')
sns.set(rc={'figure.figsize':(6, 4)})
sns.set_style('white')
sns.despine()
```

# **Data**

```python
def point():
    x = np.random.uniform(-2.5, 2)
    y = 2 * x + 5
    return x, y
```

```python
X = []
y = []


for i in range(150):
    a, b = point()
    b_noise = b + np.random.normal(0, 1)
    X.append([a])
    y.append([b_noise])
```

```python
plt.scatter(X, y, c='gold', lw=0.5, path_effects=[pe.Stroke(linewidth=2, foreground='black'), pe.Normal()])
plt.title('Data');

plt.show()
```

# **Model**

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
print('coef k:', model.coef_[0][0])
print('coef b:', model.intercept_[0])
#y = kx + b, an k and b
```

```python
prediction = model.predict(X_test)

plt.plot(X_test, prediction, linewidth=2.5, color='royalblue', label='Prediction')
plt.scatter(X_test, y_test, marker = "x", color='darkblue', label='Test data')
plt.scatter(X_train, y_train, color = 'tomato', label='Train data', path_effects=[pe.Stroke(linewidth=2, foreground='black'), pe.Normal()])

plt.xlabel('Feature')
plt.ylabel('Target')

plt.legend()
plt.title('Data');

from sklearn.metrics import mean_squared_error

print('RMSE:', mean_squared_error(prediction, y_test))
print('------------------------')
plt.show()
```
